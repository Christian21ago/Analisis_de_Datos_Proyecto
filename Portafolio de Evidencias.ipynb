```python
promStroke = dfAutos["stroke"].astype("float").mean(axis=0)
dfAutos["stroke"].replace(np.nan, promStroke, inplace=True)
dfAutos.head(5)
```

```python
promHorsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
dfAutos["horsepower"].replace(np.nan, promHorsepower, inplace=True)
dfAutos.head(5)
```

```python
promPeak = dfAutos["peak"].astype("float").mean(axis=0)
dfAutos["peak"].replace(np.nan, promPeak, inplace=True)
dfAutos.head(5)
```

<h4> 3.1 Verificar el dominio de cada una de las variable, si es necesario se ajusta</h4>

```python
dfAutos.dtypes
```

```python
dfAutos.head(5)
```

```python
dfAutos[['normalized', 'horsepower', 'peak', 'price']] = dfAutos[['normalized', 'horsepower', 'peak', 'price']].astype("int")
dfAutos[['bore', 'stroke']] = dfAutos[['bore', 'stroke']].astype("float")

dfAutos.dtypes
```

<h4> 3.2 Convertir valores en funcion de una interpretacion o contexto (tropicalizar datos)</h4>

<p>Se esta convirtiendo ciertos datos que no poseen la medicion nacional. EJ. Millas por galon a Kilometros por litro</p>

```python
dfAutos["City L/100km (calc)"] = 235/dfAutos["city"]
dfAutos["Higway L/100km (calc)"] = 235/dfAutos["higway"]
dfAutos.head(5)
```

<h4> 3.3 Normalizar valores (volver comparables valores que son asimetricos entre si)</h4>

<p>para normalizar las variables comparables hay que llevarlas entre 0 y 1</p>

```python
dfTemp = dfAutos[["horsepower", "peak"]]
dfTemp.head(3)
```

```python
dfAutos["horsepower(norm)"] = dfAutos["horsepower"]/dfAutos["horsepower"].max()
dfAutos["peak(norm)"] = dfAutos["peak"]/dfAutos["peak"].max()
dfAutos.head(5)
```

<h4> 3.4 Crear categorias o rangos (Binning)</h4>

<p>Esto ayuda a que el codigo sea mejor y consuma menos memoria RAM. Para hacer binning tiene que ser datos numericos.</p>

```python
dfTemp.head(5)
```

```python
bins = np.linspace(48, 262,4)
bins
```

```python
max(dfAutos["horsepower"])
```

```python
categoria = ["bajo", "medio", "alto"]
dfAutos["Horsepower (Bins)"] = pd.cut(dfAutos["horsepower"], bins, labels = categoria, include_lowest=True)

dfAutos[["horsepower", "Horsepower (Bins)"]].head(5)
```
